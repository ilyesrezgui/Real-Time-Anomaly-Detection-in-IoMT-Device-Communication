```python


from train import train_autoencoder


import numpy as np
import json

BASE_PATH = "/content/drive/MyDrive/IoMT-AD-Project"

X_train_s = np.load(f"{BASE_PATH}/data/intermediate/X_train_s.npy")
X_val_s   = np.load(f"{BASE_PATH}/data/intermediate/X_val_s.npy")

with open(f"{BASE_PATH}/data/intermediate/features.json") as f:
    features = json.load(f)



# Load selected features
with open(f"{BASE_PATH}/models/selected_features.json") as f:
    sel = json.load(f)

# Apply feature selection to the scaled data
X_train_fs = X_train_s[:, [features.index(f) for f in sel]]
X_val_fs   = X_val_s[:,   [features.index(f) for f in sel]]

# Train the model
history, model = train_autoencoder(
    X_train_fs,
    X_val_fs,
    base_path=BASE_PATH,
    n_features=len(sel),
    epochs=10,
    batch_size=512
)
```
