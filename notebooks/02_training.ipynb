```python
# --- First cell for all notebooks: ---
# works locally and in Google Colab 

import sys
import os


# Detect if running in Google Colab
try:
    import google.colab
    IN_COLAB = True
except ImportError:
    IN_COLAB = False

if IN_COLAB:
    # Mount Google Drive in Colab
    from google.colab import drive
    drive.mount('/content/drive')
    
    # Set BASE_PATH assuming repo is cloned in Colab
    BASE_PATH = "/content/Real-Time-Anomaly-Detection-in-IoMT-AD-Project"
else:
    # Local environment: BASE_PATH = project root (one level up from notebooks)
    BASE_PATH = os.path.abspath(os.path.join(os.getcwd(), ".."))

print(f"ğŸ“ BASE_PATH detected as: {BASE_PATH}")

# Add Spark and Models folders to Python path for imports
sys.path.append(os.path.join(BASE_PATH, "Spark"))
sys.path.append(os.path.join(BASE_PATH, "Models/Chaima"))
print("âœ… PYTHONPATH updated. Spark and Models folders included.")
```

```python
import numpy as np
import json

# Load preprocessed arrays
X_train_s = np.load(os.path.join(BASE_PATH, "Dataset/intermediate/X_train_s.npy"))
X_val_s   = np.load(os.path.join(BASE_PATH, "Dataset/intermediate/X_val_s.npy"))

# Load all features
with open(os.path.join(BASE_PATH, "Dataset/intermediate/features.json")) as f:
    features = json.load(f)

# Load selected features
with open(os.path.join(BASE_PATH, "Spark/selected_features.json")) as f:
    sel = json.load(f)

# Apply feature selection to the scaled data
X_train_fs = X_train_s[:, [features.index(f) for f in sel]]
X_val_fs   = X_val_s[:,   [features.index(f) for f in sel]]

print(f"Selected features count: {len(sel)}")
print(f"First 10 selected features: {sel[:10]}")
```

```python
from train import train_autoencoder  # train.py is inside Models/Chaima

# Train the model
history, model = train_autoencoder(
    X_train_fs,
    X_val_fs,
    base_path=BASE_PATH,
    n_features=len(sel),
    epochs=10,
    batch_size=512
)

print("âœ… Training completed. Model saved in Models folder.")
```
